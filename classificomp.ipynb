```python
import pandas as pd

d = pd.read_csv("iphone_purchase_records.csv")

X = d.iloc[:,:-1].values

y = d.iloc[:, 3].values

from sklearn.preprocessing import LabelEncoder

temp =  LabelEncoder()

X[:,0] = temp.fit_transform(X[:,0])

 

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25,random_state=0)

 

from sklearn.preprocessing import StandardScaler

sc_X = StandardScaler()

X_train = sc_X.fit_transform(X_train)

X_test = sc_X.transform(X_test)

 

#knn

from sklearn.neighbors import KNeighborsClassifier

knnc = KNeighborsClassifier(n_neighbors=5, metric="euclidean")

knnc.fit(X_train, y_train)

y_pred_knnc = knnc.predict(X_test)

 

#Lr

from sklearn.linear_model import LogisticRegression

lrc = LogisticRegression(random_state=0, solver="liblinear")

lrc.fit(X_train, y_train)

y_pred_lrc=lrc.predict(X_test)

 

#Dc

from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier(criterion = "entropy")

dtc.fit(X_train, y_train)

y_pred_dtc=dtc.predict(X_test)

 

#svm

from sklearn.svm import SVC

svm = SVC(kernel='linear')

svm.fit(X_train, y_train)

y_pred_svm=dtc.predict(X_test)

 

 

from sklearn import metrics

 

acc_knnc = metrics.accuracy_score(y_test, y_pred_knnc)

acc_lrc = metrics.accuracy_score(y_test, y_pred_lrc)

acc_dtc = metrics.accuracy_score(y_test, y_pred_dtc)

acc_svm = metrics.accuracy_score(y_test, y_pred_svm)

print("Accuracy score for KNN :",acc_knnc)

print("Accuracy score for LOG REG:",acc_lrc)

print("Accuracy score for DEC TREE :",acc_dtc)

print("Accuracy score for SVM :",acc_svm)
```
